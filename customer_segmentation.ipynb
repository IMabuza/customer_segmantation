```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
import pandas as pd
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline
```

##### Get the data

```python
df = pd.read_csv("Cust_Segmentation.csv")
df.head()
```

#### Preprocessing

```python
## Remove Address

df = df.drop('Address', axis=1)
df.head()
```

```python
## Normalise Data

from sklearn.preprocessing import StandardScaler
X = df.values[:,1:]
X = np.nan_to_num(X)
clus_dataSet = StandardScaler().fit_transform(X)
clus_dataSet
```

#### Modeling

```python
clusters = 3
k_means = KMeans(init = "k-means++", n_clusters = clusters, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

#### Insights

```python
##Addign labels to the dataset
df["clus_km"] = labels
df.head(5)
```

```python
## Average Cebtroid values by avaeraging the features in each cluster

df.groupby('clus_km').mean()
```

#### k-means will partition your customers into mutually exclusive groups, for example, into 3 clusters. The customers in each cluster are similar to each other demographically. Now we can create a profile for each group, considering the common characteristics of each cluster. For example, the 3 clusters can be:

##### AFFLUENT, EDUCATED AND OLD AGED
##### MIDDLE AGED AND MIDDLE INCOME
